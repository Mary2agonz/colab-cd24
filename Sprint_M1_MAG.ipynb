1.Diseñe e implemente una clase Producto con los siguientes atributos: id, nombre,
descripción, cantidad y precio. La clase debe incluir métodos para obtener y modificar la información de los productos.

---


5.Documente el código utilizando comentarios y docstrings apropiados para facilitar la comprensión y el mantenimiento del software.

---
Respuesta:

```
#Primero creo la clase Producto y defino los atributos requeridos en el ejercicio
class Producto:
    def __init__(self, id_producto, nombre, descripcion, cantidad, precio):
        self.id = id_producto
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

#Después genero los métodos solicitados (obtener y modificar)
#1. Método para obtener información del producto que desea saber
    def obtener_informacion(self):
        return f"\nLa información del producto es la siguiente: \nID: {self.id}\nNombre: {self.nombre}\nDescripción: {self.descripcion}\nCantidad: {self.cantidad}\nPrecio: ${self.precio}"

#2. Método para modificar atributo y agregar nuevo valor a ese atributo de un id particular de la base de datos
    def modificar_atributo(self, atributo, nuevo_valor):
        if atributo == "nombre":
            self.nombre = nuevo_valor
        elif atributo == "descripcion":
            self.descripcion = nuevo_valor
        elif atributo == "cantidad":
            self.cantidad = nuevo_valor
        elif atributo == "precio":
            self.precio = nuevo_valor
        #agregué este else por si el atributo que ingrese el usuario no existe en la base de datos, le envíe un mensaje de error
        else:
            print("\nAtributo inválido. \nAsegurese de escribir el atributo con minúscula.")

#Ahora creo una base de datos con productos ejemplo
productos = [Producto(1, "Poleron", "Poleron de poliéster", 15, 28790),
            Producto(2, "Chaqueta", "Chaqueta impermeable", 10, 69990),
            Producto(3, "Pantalón", "Pantalon de secado rápido", 20, 22590),
            Producto(4, "Zapatos", "Zapatos de trekking", 5, 89990),
            Producto(5, "Calcetines", "Calcetines de cobre", 40, 7990)]

#Ahora defino la función para mostrar el menú de opciones del programa
def mostrar_menu():
    print("\n---- MENÚ ----")
    print("1. Consultar información de un producto")
    print("2. Modificar información de un producto")
    print("3. Salir")

#Creo el bucle principal del programa (utilizando while y if-elif-else) que comienza por mostrar el menú de opciones
while True:
    mostrar_menu()
    #aquí le pido al usuario que elija una de las 3 opciones propuestas
    opcion = input("\nSeleccione una opción: ")

#Opción 1 para seleccionar el id del producto que quiere saber la info
    if opcion == "1":
        #primero solicito al usuario que ingrese el id del producto a consultar
        id_producto = int(input("Ingrese el ID del producto que desea consultar: "))
        #agregué esta restricción si el id ingresado no está en la base de datos
        id_ingresado = False
        #además agregué este bloque for para que recorra la lista de productos y verifique que el id ingresado está en la lista de productos
        for producto in productos:
            if producto.id == id_producto:
                id_ingresado = True
                #si el id está en la base de datos, entonces imprimo la info del producto consultado y termino el bucle
                print(producto.obtener_informacion())
                break
        #finalmente imprimo el mensaje de error si el id no se encuentra en la base de datos
        if not id_ingresado:
            print("\nEl producto ingresado no está en la base de datos.")

#Opción 2 para seleccionar el producto, mediante el id, que quiere modificar
    elif opcion == "2":
        #primero solicito al usuario que ingrese el id del producto a modificar
        id_producto = int(input("Ingrese el ID del producto que desea modificar: "))
        #agregué esta restricción para evitar que el programa continúe con la modificación del producto si el id no está en la base de datos
        id_ingresado = False
        #además agregué este bloque for para que recorra la lista de productos y verifique que el id ingresado está en la lista de productos
        for producto in productos:
            if producto.id == id_producto:
               id_ingresado = True
               #si se encuentra el id en nuestra lista, entonces solicitamos al usuario el atributo a modificar
               atributo = input("Ingrese el atributo que desea modificar (nombre, descripcion, cantidad, precio): ")
               #una vez ingresado el atributo a modificar solicito el nuevo valor para ese atributo
               nuevo_valor = input("Ingrese el nuevo valor: ")
               #llamo a la función para modificar la info del producto
               modificar_producto(id_producto, atributo, nuevo_valor)
               #cuando finaliza la modificación le envío un mensaje de exito al usuario y cierro el bucle
               print("\nProducto modificado exitosamente.")
               break
        #finalmente imprimo el mensaje de error si el id no se encuentra en la base de datos
        if not id_ingresado:
            print("\nEl producto ingresado no está en la base de datos.")

#Opción 3 para salir del bucle while y terminar el programa
    elif opcion == "3":
        print("\nEl programa finalizó. Adiós!")
        break

#Finalmente agrego el else por si el usuario digita una opción que no está en el menú
    else:
        print("\nOpción no válida. Por favor, seleccione una opción válida.")
```

2.Diseñe e implemente una clase Inventario que utilice una estructura de datos (por ejemplo, un diccionario) para almacenar objetos de la clase Producto. La clase Inventario debe incluir métodos para realizar las siguientes operaciones:
• Agregar un producto al inventario
• Eliminar un producto del inventario
• Actualizar la información de un producto en el inventario
• Buscar un producto en el inventario por su id
• Listar todos los productos en el inventario, incluyendo información como la cantidad total de productos y el valor total del inventario

---


3.Implemente una menú de usuario que permita a los usuarios interactuar con el
sistema de gestión de inventario. El menú de incluir opciones para realizar las diferentes operaciones disponibles en la clase Inventario. Utilice un bucle while y estructuras de control de flujo if-elif-else para manejar las diferentes opcionesdel menú. (opcional)

---


4.Asegúrese de que el sistema maneje adecuadamente errores y excepciones, como entradas inválidas del usuario o intentos de actualizar o eliminar productos que no existen en el inventario.


---


5.Documente el código utilizando comentarios y docstrings apropiados para facilitar la comprensión y el mantenimiento del software.

---
Respuesta:

```
#Para el desarrollo de las actividades 2-3-4 reutilicé varias líneas y comentarios de mi código anterior (actividad N°1)
#Para este nuevo código no propuse una base de datos ejemplo, para almacenar los productos a ingresar uno a uno
#Primero creo la clase Producto y sus atributos
class Producto:
    def __init__(self, id_producto, nombre, descripcion, cantidad, precio):
        self.id = id_producto
        self.nombre = nombre
        self.descripcion = descripcion
        self.cantidad = cantidad
        self.precio = precio

#Después genero el método para obtener información del producto que desea saber
    def obtener_informacion(self):
        return f"\nLa información del producto es la siguiente: \nID: {self.id}\nNombre: {self.nombre}\nDescripción: {self.descripcion}\nCantidad: {self.cantidad}\nPrecio: ${self.precio}"

#Creo la clase Inventario y la inicio para almacenar productos
class Inventario:
    def __init__(self):
        self.productos = {}

### Método para agregar un nuevo producto al inventario
    def agregar_producto(self, producto):
        self.productos[producto.id] = producto
        print("\nEl producto fue ingresado con exito.")

### Método para eliminar un producto del inventario dado su ID
    def eliminar_producto(self, id_producto):
        if id_producto in self.productos:
            del self.productos[id_producto]
            print("\nEl producto fue eliminado del inventario.")
        else:
            print("\nEl producto ingresado no está en el inventario.")

### Método para actualizar la información de un producto en el inventario
    def actualizar_producto(self, id_producto, atributo, nuevo_valor):
        if id_producto in self.productos:
            producto = self.productos[id_producto]
            if hasattr(producto, atributo):
                setattr(producto, atributo, nuevo_valor)
                print("\nInformación del producto actualizada con exito")
            else:
                print("\nAtributo inválido.")
        else:
            print("\nEl producto ingresado no está en el inventario.")

### Método para buscar un producto en el inventario por su ID
    def buscar_producto(self, id_producto):
        if id_producto in self.productos:
            print("\nSe encontró el producto consultado")
            producto = self.productos[id_producto]
            print(producto.obtener_informacion())
        else:
            print("\nEl producto ingresado no está en el inventario.")

### Método para listar todos los productos en el inventario
    def listar_productos(self):
        print("\nA continuación se muestran todos los productos del inventario:\n")
        for producto in self.productos.values():
            print(producto.obtener_informacion())

### Método para obtener la cantidad total de productos en el inventario
    def cantidad_total_productos(self):
        return len(self.productos)

### Método para calcular el valor total del inventario (suma del precio de todos los productos)
    def valor_total_inventario(self):
        total = sum(producto.cantidad * producto.precio for producto in self.productos.values())
        return total


def mostrar_menu():
### Función para mostrar el menú de opciones al usuario
    print("\n---- MENÚ ----")
    print("1. Agregar producto")
    print("2. Eliminar producto")
    print("3. Actualizar información de un producto")
    print("4. Buscar producto por ID")
    print("5. Listar todos los productos")
    print("6. Mostrar cantidad total de productos")
    print("7. Mostrar valor total del inventario")
    print("8. Salir")


inventario = Inventario()

#Creo el bucle principal del programa (utilizando while y if-elif-else) que comienza por mostrar el menú de opciones
while True:
    mostrar_menu()
    opcion = input("\nSeleccione una opción: ")

### Opción para agregar un producto al inventario
    if opcion == "1":
        id_producto = int(input("Ingrese el ID del nuevo producto: "))
        nombre = input("Ingrese el nombre del producto: ")
        descripcion = input("Ingrese la descripción del producto: ")
        cantidad = int(input("Ingrese la cantidad del producto: "))
        precio = float(input("Ingrese el precio del producto: "))
        producto_nuevo = Producto(id_producto, nombre, descripcion, cantidad, precio)
        inventario.agregar_producto(producto_nuevo)

### Opción para eliminar un producto del inventario
    elif opcion == "2":
        id_producto = int(input("Ingrese el ID del producto a eliminar: "))
        inventario.eliminar_producto(id_producto)

### Opción para actualizar la información de un producto en el inventario
    elif opcion == "3":
        id_producto = int(input("Ingrese el ID del producto a actualizar: "))
        atributo = input("Ingrese el atributo a actualizar (nombre, descripcion, cantidad, precio): ")
        nuevo_valor = input("Ingrese el nuevo valor: ")
        inventario.actualizar_producto(id_producto, atributo, nuevo_valor)

### Opción para buscar un producto en el inventario por su ID
    elif opcion == "4":
        id_producto = int(input("Ingrese el ID del producto a buscar: "))
        inventario.buscar_producto(id_producto)

### Opción para listar todos los productos en el inventario
    elif opcion == "5":
        inventario.listar_productos()

### Opción para mostrar la cantidad total de productos en el inventario
    elif opcion == "6":
        print("\nCantidad total de productos:", inventario.cantidad_total_productos())

### Opción para mostrar el valor total del inventario
    elif opcion == "7":
        print("\nValor total del inventario:", inventario.valor_total_inventario())

#Opción 8 para salir del bucle while y terminar el programa
    elif opcion == "8":
        print("\nEl programa finalizó. Adiós!")
        break

#Finalmente agrego el else por si el usuario digita una opción que no está en el menú
    else:
        print("\nOpción no válida. Por favor, seleccione una opción válida.")
```
