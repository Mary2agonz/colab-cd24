* En la primera iteración se asigna a cada punto el centroide más cercano (C1 o C2) mendiante el calculo de la distancia euclidiana entre los puntos, es decir la distancia entre cada punto y los centroides predefinidos. 5. Enuncie al menos tres diferencias entre el análisis factorial y el método
de componentes principales.

...................................

Respuesta:



```
(1) El análisis factorial se centra en identificar variables latentes relacionadas con la causa de un problema, mientras que el método de componentes principales busca reducir la dimensionalidad conservando la variabilidad y destacando las relaciones más importantes entre las variables observadas.
(2) El análisis factorial supone la existencia de variables latentes que son explicadas por las variables observadas, mientras que el modelo de componentes principales no asume la existencia de variables latentes y solo considera las variables que componen el modelo.
(3) En el análisis factorial, la interpretación se basa en la comprensión teórica de las variables latentes, mientras que en el modelo de componentes principales, la interpretación se realiza a través de las combinaciones lineales de las variables sin una interpretación directa de cada componente.
```

#Actividad N°1

**Búsqueda de Anomalías**: Usando las columnas de comportamiento de
compra, encuentre registros anómalos correspondientes al 5% de la muestra,
y descártelos

````python
#Primero importo las librerías necesarias
#Además de pandas utilizo IsolationForest de sklearn para identificar puntos anómalios
import pandas as pd
from sklearn.ensemble import IsolationForest

#Cargo los datos desde link drive
link = "https://drive.google.com/uc?id=1hSiLBfvMi051RJM18_yxs3bOBokCTToq"
#Creo el DataFrame con los datos cargados
df = pd.read_csv(link)

#Selecciono las columnas relevantes para la detección de anomalías
#Omito la seleccion de MntFruits ya que no existe en la base de datos entregada para el ejercicio
columnas = ['MntWines', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

#Creo un nuevo dataframe con las columnas relevantes
df_col = df[columnas]

#Aplico la función isolationforest() para la detección de anomalías (Isolation Forest)
#Defino el porcentaje esperado de puntos anomalos (en este caso 5%)
#Defino un valor específico para random_state para asegurar que el modelo produzca la misma cantidad de resultados cada vez que se ejecute el modelo
deteccion = IsolationForest(contamination=0.05, random_state=20)
#Ajusto el modelo a los datos de entrenamiento
deteccion.fit(df_col)

#Identifico los registros anómalos mediante el uso del método predict() sobre nuestro modelo de detección
anomalias = deteccion.predict(df_col)

#Realizo el filtro de los registros anómalos y los elimino del conjunto de datos original
#Con "anolaias == 1" genero una matriz booleana donde 1 es True, osea no anómalo, y False anómalo
#Por lo tanto, este nuevo df (df_depurado) retiene solo los registros no anómalos
df_depurado = df[anomalias == 1]

#Finalmente imprimo las primeras 20 filas del df original seguido del depurado para que se vean las filas que fueron excluidas por presentar registros anómalos
print("DataFrame con las columnas originales:")
display(df.head(20))
print("..........................................................................")
print("DataFrame post depuración de registros anómalos:")
display(df_depurado.head(20))
````

#Actividad N°2

**Feature Extraction**: Usando los datos filtrados reduzca el número de
variables. Busque una representación que le permita explicar un 90% de la
varianza original de los datos.

````python
#Importo las nuevas librerias necesarias para lo que pide el ejercicio
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

"""
#Primero importo las librerías necesarias
#Además de pandas utilizo IsolationForest de sklearn para identificar puntos anómalios
import pandas as pd
from sklearn.ensemble import IsolationForest

#Cargo los datos desde link drive
link = "https://drive.google.com/uc?id=1hSiLBfvMi051RJM18_yxs3bOBokCTToq"
#Creo el DataFrame con los datos cargados
df = pd.read_csv(link)

#Selecciono las columnas relevantes para la detección de anomalías
#Omito la seleccion de MntFruits ya que no existe en la base de datos entregada para el ejercicio
columnas = ['MntWines', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

#Creo un nuevo dataframe con las columnas relevantes
df_col = df[columnas]
"""

#Estandarizo los datos
standar = StandardScaler()
#Ejecuto la estadarización sobre las columnas filtradas para reducir el número de variables
df_standar = standar.fit_transform(df_col)

#Aplico el PCA con todos los componentes
pca = PCA()
df_pca = pca.fit_transform(df_standar)
#Imrpimo el df de los datos pca transformados
#print(df_pca)

#Grafico la proporción acumulativa de la varianza explicada
plt.figure(figsize=(7, 5))
#aplico "explained_variance_ratio_" para obtener la proporcion de la varianza explicada por cada componente
#ademas, creo una secuencia de números para que el eje x comience en 1 en vez de 0 (esto lo corregi despues de plotear sin esto y ver que empezaba en componente 0 y no correspondía)
plt.plot(range(1, len(pca.explained_variance_ratio_) + 1),
                  #Con esta línea se calcula la varianza acumulada por componente principal
                  np.cumsum(pca.explained_variance_ratio_),
                  #esta linea define que la linea de varianza se continua y con puntos para los valores
                  marker='o', linestyle='-', color='black')
#Defino el nombre del titulo y de los ejes del gráfico
plt.xlabel('Número de componentes principales')
plt.ylabel('Varianza acumulada explicada')
plt.title('Varianza explicada por componentes principales')
#Defino que se genere una linea horizontal al 90% de la varianza explicada es decir al 0.9
plt.axhline(y=0.90, color='red', linestyle='-.', label='90% de varianza explicada')
plt.legend()
#Disminuyo la visibilidad de la grilla de fondo para no saturar la visibilidad del gráfico
plt.grid(True, alpha=0.3)
#Imprimo el gráfico y el mensaje asociado al resultado del ejercicio
plt.show()

#Finalmente aplico la siguiente linea para encontrar el número de componentes que se necesitan para explicar al menos el 90% de la varianza
n_components_90 = np.argmax(np.cumsum(pca.explained_variance_ratio_) >= 0.90) + 1
#Imrpimo el resultado
print(f"\nEl 90% de la varianza total se explica con, al menos, los primeros {n_components_90} componentes principales del PCA")
````

#Actividad N°3

**Clustering**:

a) Usando los resultados anteriores clusterice el comportamiento de sus
clientes. El área de marketing solicitó que el número de clusters sea no
mayor a 8. Encuentre el número óptimo de clusters considerando esa
restricción

b) Reporte los centroides de cada cluster en términos de las variables
originales. Interprete.



````python
#Añado la librería necesaria faltante además de las que ya fueron cargadas
from sklearn.cluster import KMeans

#Para este ejercicio utilizaré el df_depurado ya que es el que incluye los id y puedo analizar el comportamiento de los clientes de mejor forma
#df_depurado.info()

#Antes de la clusterización, elimino  columna ID de nuestro df mediante el uso del método drop()
df_cluster = df_depurado.drop(columns=['ID'])

#Luego, defino el número máximo de clusters (en este caso máximo 8)
max_clusters = 8

#Creo una lista vacía para almacenar los valores de inercia
inercia = []

#Creo el bucle for para iterar sobre cada cluster y calcular la inercia de cada uno
for n_clusters in range(1, max_clusters+1):
    #Creo la instancia kmeans para regenar números aleatorios sobre la función Kmeans y garantizar que se reproduzcan la misma cantidad de datos cada vez que se ejecuta la linea
    kmeans = KMeans(n_clusters=n_clusters, random_state=20)
    #Aplico el kmeans al df que tiene los datos filtrados y depurados
    kmeans.fit(df_cluster)
    #Almaceno la inercia
    inercia.append(kmeans.inertia_)

#Para encontrar el número óptimo de clusters utilizo el método del codo que grafica la inercia y donde se flectala curva es el numero optimo de clusters
plt.plot(range(1, max_clusters+1), inercia, marker='o', linestyle='-', color='purple')
#Defino el título y nombre de ejes
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.title("Número óptimo de clusters")
#Imrpimo el gráfico
plt.show()
````

````python
#Ahora que tengo el gráfico, defino el número óptimo de clusters basado en lo observado en el gráfico
#Que en este caso se flecta más evidentemente en 3
n_clusters_optimo = 3

#Ajsuto el modelo KMeans al número óptimo de clusters
kmeans_optimo = KMeans(n_clusters=n_clusters_optimo, random_state=20)
kmeans_optimo.fit(df_pca)

#Calculo los centroides de cada cluster con respecto a los componentes principales
centroides_pca = kmeans_optimo.cluster_centers_

#Calculo los centroides de cada cluster con respecto a las variables originales
#Para ello invierto la transformación PCA
centroides_original = standar.inverse_transform(pca.inverse_transform(centroides_pca))

#Creo un DataFrame para los centroides con respecto a las variables originales
df_centroides = pd.DataFrame(centroides_original, columns=df_col.columns)

#Finalmente imprimo los centroides de cada cluster
print("Centroides de cada cluster en términos de las variables originales:")
display(df_centroides)
````
