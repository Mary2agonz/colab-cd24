5. Enuncie al menos tres diferencias entre el análisis factorial y el método
de componentes principales.

...................................

Respuesta:



```
(1) El análisis factorial se centra en identificar variables latentes relacionadas con la causa de un problema, mientras que el método de componentes principales busca reducir la dimensionalidad conservando la variabilidad y destacando las relaciones más importantes entre las variables observadas.
(2) El análisis factorial supone la existencia de variables latentes que son explicadas por las variables observadas, mientras que el modelo de componentes principales no asume la existencia de variables latentes y solo considera las variables que componen el modelo.
(3) En el análisis factorial, la interpretación se basa en la comprensión teórica de las variables latentes, mientras que en el modelo de componentes principales, la interpretación se realiza a través de las combinaciones lineales de las variables sin una interpretación directa de cada componente.
```

Suponga que tenemos una empresa, cuya área de marketing, desea generar
una clusterización para poder tener mayor éxito con sus campañas. Para esto
nos entregó una base de datos (llamada evaluación_mkt_campaign.csv) que
contiene los siguientes campos:
1. ID: Identificador único de cliente
2. Age: Edad del cliente
3. Seniority: Días desde el enrolamiento del cliente
4. Children: Número hijos

7. MntWines: Gasto en vino en los últimos 2 años

9. MntMeatProducts: Gasto en carne en los últimos 2 años
10. MntFishProducts: Gasto en pescado en los últimos 2 años
11. MntSweetProducts: Gasto en dulces en los últimos 2 años
12. MntGoldProds: Gasto en oro en los últimos 2 años
13. NumDealsPurchases: Número de compras realizada bajo descuento
14. NumWebPurchases: Número de compras realizadas por la web
15. NumCatalogPurchases: Número de compras realizadas por catálogo
16. NumStorePurchases: Número de compras realizadas en la tienda
17. NumWebVisitsMonth: Número de visitas a la página web


En base a estos datos, realice lo siguiente:


#Actividad N°1

**Búsqueda de Anomalías**: Usando las columnas de comportamiento de
compra, encuentre registros anómalos correspondientes al 5% de la muestra,
y descártelos

````python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Cargar los datos
data = pd.read_csv("evaluación_mkt_campaign.csv")

# Seleccionar columnas relevantes para la detección de anomalías
columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Crear un nuevo dataframe con las columnas relevantes
purchase_data = data[columns]

# Entrenar el modelo de detección de anomalías (Isolation Forest)
clf = IsolationForest(contamination=0.05, random_state=42)
clf.fit(purchase_data)

# Identificar registros anómalos
anomalies = clf.predict(purchase_data)

# Filtrar los registros anómalos y eliminarlos del conjunto de datos original
cleaned_data = data[anomalies == 1]

# Guardar el nuevo conjunto de datos sin registros anómalos
cleaned_data.to_csv("evaluación_mkt_campaign_cleaned.csv", index=False)
````

#Actividad N°2

**Feature Extraction**: Usando los datos filtrados reduzca el número de
variables. Busque una representación que le permita explicar un 90% de la
varianza original de los datos.

````python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Cargar los datos filtrados
data_filtered = pd.read_csv("evaluación_mkt_campaign_cleaned.csv")

# Seleccionar solo las columnas relevantes para el PCA (las relacionadas con el gasto en productos)
columns = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
data_pca = data_filtered[columns]

# Estandarizar los datos
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data_pca)

# Aplicar PCA
pca = PCA(n_components=0.9)  # Conservar el 90% de la varianza original
data_pca_transformed = pca.fit_transform(data_scaled)

# Crear un nuevo DataFrame con las componentes principales
columns_pca = [f"Componente_{i+1}" for i in range(data_pca_transformed.shape[1])]
data_pca_transformed_df = pd.DataFrame(data_pca_transformed, columns=columns_pca)

# Concatenar el ID de cliente al DataFrame de las componentes principales
data_pca_transformed_df['ID'] = data_filtered['ID']

# Guardar el nuevo conjunto de datos con las componentes principales
data_pca_transformed_df.to_csv("evaluación_mkt_campaign_pca.csv", index=False)
````

#Actividad N°3

**Clustering**:

a) Usando los resultados anteriores clusterice el comportamiento de sus
clientes. El área de marketing solicitó que el número de clusters sea no
mayor a 8. Encuentre el número óptimo de clusters considerando esa
restricción

b) Reporte los centroides de cada cluster en términos de las variables
originales. Interprete.



````python
import pandas as pd
from sklearn.cluster import KMeans
import numpy as np

# Cargar los datos con las componentes principales
data_pca = pd.read_csv("evaluación_mkt_campaign_pca.csv")

# Eliminar la columna de ID de cliente antes de la clusterización
data_pca.drop(columns=['ID'], inplace=True)

# Definir el número máximo de clusters permitidos
max_clusters = 8

# Inicializar una lista para almacenar los valores de inercia
inertia_values = []

# Probar diferentes números de clusters y calcular la inercia para cada uno
for n_clusters in range(1, max_clusters+1):
    kmeans = KMeans(n_clusters=n_clusters, random_state=42)
    kmeans.fit(data_pca)
    inertia_values.append(kmeans.inertia_)

# Encontrar el codo en la curva de inercia para determinar el número óptimo de clusters
optimal_clusters = np.argmin(np.gradient(inertia_values)) + 1

# Imprimir el número óptimo de clusters encontrado
print("Número óptimo de clusters:", optimal_clusters)

# Clusterización con el número óptimo de clusters
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
data_pca['Cluster'] = kmeans.fit_predict(data_pca)

# Calcular los centroides de cada cluster en términos de las componentes principales
centroids_pca = kmeans.cluster_centers_

# Invertir la transformación PCA para obtener los centroides en términos de las variables originales
centroids_original = scaler.inverse_transform(pca.inverse_transform(centroids_pca))

# Crear un DataFrame para los centroides en términos de las variables originales
centroid_df = pd.DataFrame(centroids_original, columns=data_filtered.columns[1:])

# Imprimir los centroides
print("\nCentroides de cada cluster en términos de las variables originales:")
print(centroid_df)
````
