#Descripción de la Actividad Evaluativa

---

Vamos a crear un sistema de facturación y bodega simplificado, el cual tendrá diferentes opciones:

a) Ingresar producto: debe ingresarse el nombre y el precio del producto.
Tendrá un sub menú: nuevo producto, guardar archivo o volver.

b) Ingresar factura: debe ingresarse el nombre del cliente, luego deberá generar un submenú: ingresar producto, guardar y salir.

Los productos se ingresan con nombre y precio. La factura además debe guardar la fecha y hora actual de la venta.

c) Ver resumen: en esta opción se debe desplegar: Cantidad de ventas y monto total vendido. Opción de guardar resumen (ingresando nombre de archivo).

d) Ver detalle: se debe tener las opciones desplegar: ventas por usuario, ventas por producto, ventas diarias, guardar vista (ingresando el nombre de archivo).

e) Borrar datos: elimina los archivos guardados (elegir productos, facturas, todos o volver)

f) Salir: termina la ejecución del programa.

```
#Antes que todo importo las librerías
import pandas as pd
from datetime import datetime

#Importadas las librerías, defino los nombres de los archivos que serán cargados
archivo_productos = 'productos.csv'
archivo_facturas = 'facturas.csv'
archivo_resumen = 'resumen.csv'
archivo_detalle = 'detalle.csv'

#Ahora cargo los archivos y creo los dataframe
try:
    df_productos = pd.read_csv(archivo_productos)
#Agrego la excepción FileNotFoundError para que el programa no se detenga
#Esta excepción permite manejar el error en el que el archivo no se encuentra en el sistema
except FileNotFoundError:
#Si el archivo no se encuentra en el sistema entonces se creará un nuevo DataFrame vacío para el archivo "productos", con las columnas que definimos
    df_productos = pd.DataFrame(columns=['Nombre', 'Precio'])

try:
    df_facturas = pd.read_csv(archivo_facturas)
#Al igual que para productos, agregué la excepción FileNotFoundError para que el programa no se detenga y maneje el error que el archivo no está en el sistema
except FileNotFoundError:
#Si el archivo no se encuentra en el sistema entonces se creará un nuevo DataFrame vacío para el archivo "facturas", con las columnas que definimos
    df_facturas = pd.DataFrame(columns=['Cliente', 'Fecha', 'Productos'])

#-----------------------------------------------------------------------------------------
#A) Defino el método para que el cliente ingrese los atributos del producto
def ingresar_producto():
#Agrego la palabra clave "global" para indicar que la variable "df_productos" se refiere a la variable global fuera de la función
#De esta manera cualquier cambio realizado en df_productos dentro de la función modificará a la variable global df_productos
    global df_productos
    nombre = input("Ingrese el nombre del producto: ")
    while True:
        precio_input = input("Ingrese el precio del producto: ")
#Agrego el bloque try-except para manejar el tipo de dato que queremos guardar en nuestra columna precio
        try:
            precio = float(precio_input)
            break
#Agrego la excepción ValueError para manejar el error en el que el usuario ingrese un dato no válido (un valor que no se puede convertir a decimal)
        except ValueError:
#Si resulta que el usuario ingreso un valor inválido, le imprimo el mensaje de error
            print("\nError: Ingrese un valor numérico válido.")

#Una vez que el usuario ingresa un valor válido, el bucle se detiene y el programa continúa con el ingreso del producto
    nuevo_producto = pd.DataFrame({'Nombre': [nombre], 'Precio': [precio]})
#Agregué la función pd.concat() para unir objetos de datos de los df de productos
#Además agregué el ignore_index=True para que se ignoren los índices existentes en cada df, y se cree un nuevo índice para el df resultante
    df_productos = pd.concat([df_productos, nuevo_producto], ignore_index=True)
#Una vez finalizado el proceso anterior, se imprime el mensaje de exito
    print("\nProducto ingresado con éxito.")

#-----------------------------------------------------------------------------------------
#B) INGRESAR FACTURA
#Creo el método para ingresar el nombre del cliente y el submenu (ingresar, guardar, salir)
def ingresar_factura():
#Agrego la palabra clave "global" para indicar que la variable "df_facturas" se refiere a la variable global fuera de la función
#De esta manera cualquier cambio realizado en df_facturas dentro de la función modificará a la variable global df_facturas
    global df_facturas
    cliente = input("Ingrese el nombre del cliente: ")
#Ingresado el nombre del cliente, se agrega el método strftime sobre la función datetime para obtener la fecha y hora del sistema convertidas a una cadena de texto
#En este caso, el formato elegido es año-mes-día hora:minuto:segundos
    fecha = datetime.now().strftime('%Y-%m-%d %H:%M:%S')

#Ahora creo una lista vacía que almacenará los nombres de los productos que se ingresen en la factura
    productos_factura = []
    while True:
        nombre_producto = input("Ingrese el nombre del producto (o 'x' para salir): ")
#Agrego esta condición con la función lower() para que el x siempre se convierta a minúscula y continúe ejecutandose el programa
        if nombre_producto.lower() == 'x':
            #Agrego esta condicional para que si hay productos en la factura salimos del bucle con break
            if productos_factura:
                break
#Agrego else por si la condición anterior no se cumple, y si no ha ingresado la X, por lo tanto le imprimo un mensaje de error
            else:
                print("Ha ocurrido un error: Debe ingresar al menos un producto.")
                #Finalmente agrego continue para que el bucle vuelva a comenzar, es decir vuelve a solicitar al usuario que ingrese el producto hasta que ingrese la x de salir
                continue

#El uso del método append() permite almacenear los productos ingresados en la lista productos_factura
        productos_factura.append(nombre_producto)

#Despues de ingresada la factura, el bucle while se detiene y se crea un nuevo df llamado "nueva_factura"
#Este nuevo df se crea con 3 columnas con los valores derivados de las variables cliente y fecha, y de la lista productos_factura
    nueva_factura = pd.DataFrame({'Cliente': [cliente], 'Fecha': [fecha], 'Productos': [productos_factura]})
#Luego, se une este nuevo df con el df_facturas, mediante el uso de la función pd.concat()
#Además agregué el ignore_index=True para que se ignoren los índices existentes en cada df, y se cree un nuevo índice para el df resultante
    df_facturas = pd.concat([df_facturas, nueva_factura], ignore_index=True)
#Una vez finalizado el proceso anterior, se imprime el mensaje de éxito
    print("\nFactura ingresada con éxito.")

#-----------------------------------------------------------------------------------------
#C) Método para ver el resumen
def ver_resumen():
#Utilizo la palabra clave global para que las modificaciones posteriores se realicen sobre las variables globales que se utilizarán dentro de las funciones siguientes
    global df_facturas, df_productos
#Agregué la función len() para calcular el número de filas del df_facturas y así tener la cantidad de ventas registradas
    cantidad_ventas = len(df_facturas)
#Para calcular el monto total vendido se aplicó la función apply lambda row para que la operación se ejecute a cada fila
#Además, dentro de la función lambda se está filtrando los productos cuyos nombres se encuentran dentro de la lista de productos de la fila Productos
#Además, agrego axis=1 para que la función apply se ejecute a lo largo de las filas del df
#También se incluye la función isin() para verificar si cada producto en la fila está presente en el df de productos
#Finalmente se extraen los precios de los productos y se suman con la función .sum() que suma el resultado de todas las filas
    monto_total_vendido = df_facturas.apply(lambda row: sum(df_productos[df_productos['Nombre'].isin(row['Productos'])]['Precio']), axis=1).sum()

#Ejecutado el monto total vendido se imprime el resumen mediante los siguientes print:
    print(f"\nResumen: ")
    print(f"Cantidad de ventas: {cantidad_ventas}")
    print(f"Monto total vendido: {monto_total_vendido}")

#Las siguientes líneas se agregan para que se almacene el resumen si el usuario quiere guardarlo
    opcion_guardar = input("\n¿Desea guardar el resumen? (s/n): ")
#Con este if se verifica si la entrada del usuario, convertida en minúscula mediante lower(), es igual a "s" para guardar el resumen
    if opcion_guardar.lower() == 's':
#Si el usuario quiere guardar el resumen, aquí se le solicita que ingrese el nombre del archivo a guardar
        archivo_guardar = input("\nIngrese el nombre del archivo para guardar el resumen: ")
#Luego se crea un nuevo df (df_resumen) que contiene dos columnas "Cantidad_Ventas" y "Monto_Total_Vendido"
        df_resumen = pd.DataFrame({'Cantidad_Ventas': [cantidad_ventas], 'Monto_Total_Vendido': [monto_total_vendido]})
#Después se guarda el df en un archivo csv con el nombre designado por el usuario
#Se agrega el argumento index=False para evitar que se incluyan los índices de fila en el archivo a exportar en csv
        df_resumen.to_csv(archivo_guardar + '.csv', index=False)
#Finalmente se imprime el mensaje que el archivo se ha guardado correctamente
        print(f"\nResumen guardado con éxito en {archivo_guardar}.csv")

#-----------------------------------------------------------------------------------------
#D) Método para ver el detalle
def ver_detalle():
#Agrego la palabra clave "global" para indicar que la variable "df_facturas" se refiere a la variable global fuera de la función
#De esta manera cualquier cambio realizado en df_facturas dentro de la función modificará a la variable global df_facturas
    global df_facturas
    print("\nDetalle:")
    print("1. Ventas por usuario")
    print("2. Ventas por producto")
    print("3. Ventas diarias")
    print("4. Volver")

    opcion_detalle = input("\nSeleccione una opción: ")
#Ahora desglozo el submenú para ver detalle
    if opcion_detalle == '1':
#Agrupo el df_factura por la columna cliente y uso size() para contar la frecuencia de cada valor único en el df agrupado
#Seguido, reinicio el índice con reset_index y renombro la columna resultante como "Cantidad_Ventas"
        ventas_usuario = df_facturas.groupby('Cliente').size().reset_index(name='Cantidad_Ventas')
        print(ventas_usuario)
    elif opcion_detalle == '2':
#Uso la función explode() para expandir la columna Productos en filas separadas
        ventas_producto = df_facturas.explode('Productos')
#Luego agrupo el df por la columna Productos y al igual que la opción anterior, uso size() y reset_index para contar la frecuencia y renombrar la columna resultante
        ventas_producto = ventas_producto.groupby('Productos').size().reset_index(name='Cantidad_Ventas')
        print(ventas_producto)
    elif opcion_detalle == '3':
#Para la opción 3 agrupo el df por la columna fecha utilizando str.spplit() para dividir la cadena en fecha y hora por eso se dejo el espacio ' '
#Además se agrega el srt[0] para seleccionar el primer elemento de cada lista y después aplicar el método str.plit()
#Al igula que en las opciones anteriores utilizo el size() para contar el número de ventas, en este caso, para cada fecha, y seguidamente reinicio el indice y renombro la columna resultante
        ventas_diarias = df_facturas.groupby(df_facturas['Fecha'].str.split(' ').str[0]).size().reset_index(name='Cantidad_Ventas')
        print(ventas_diarias)
    elif opcion_detalle == '4':
#Esta opción retorna al método ver_detalle() para volver al submenú d) ver detalle
        return

#Posteriormente, se le consulta al usuario si desea guardar la vista
    opcion_guardar = input("\n¿Desea guardar la vista? (s/n): ")
#Agrego la función lower() para que en caso de querer guardar, el 's' siempre se convierta a minúscula y continúe ejecutandose el programa
    if opcion_guardar.lower() == 's':
#Si efectivamente desea guardar la vista, le solicito el nombre del archivo a guardar, dependiendo de la opción del submenú a continuación:
        archivo_guardar = input("\nIngrese el nombre del archivo para guardar la vista: ")
#Independiente de la opción que seleccione, usamos el to_csv para que el archivo se guardará como un csv, y index=False para que no se incluya el indice en el archivo de salida
        if opcion_detalle == '1':
            ventas_usuario.to_csv(archivo_guardar + '.csv', index=False)
        elif opcion_detalle == '2':
            ventas_producto.to_csv(archivo_guardar + '.csv', index=False)
        elif opcion_detalle == '3':
            ventas_diarias.to_csv(archivo_guardar + '.csv', index=False)
#Finalmente se imprime el mensaje que el archivo se ha guardado correctamente
        print(f"\nVista guardada con éxito en {archivo_guardar}.csv")
#Si apreta 'n' en respuesta a la pregunta de desea guardar la vista, y/o cualquier otra letra, el programa simplemente continuará ejecutandose sin guardar la vista

#-----------------------------------------------------------------------------------------
#E) Método para borrar datos
def borrar_datos():
#Utilizo la palabra clave global para que las modificaciones posteriores se realicen sobre las variables globales que se utilizarán dentro de las funciones siguientes
    global df_productos, df_facturas
#Imprimo el submenú para borrar datos
    print("\nBorrar datos: ")
    print("1. Borrar productos")
    print("2. Borrar facturas")
    print("3. Borrar todos los datos")
    print("4. Volver")

#Luego de presentar el submenú, le solicito al usuario que elija una opción la cual se almacena en la variable opcion_borrar
    opcion_borrar = input("\nSeleccione una opción: ")

    if opcion_borrar == '1':
#En esta opción se crea un df vacío con las columnas nombre y precio, que sobreescribe cualquier df con datos de productos ingresados
        df_productos = pd.DataFrame(columns=['Nombre', 'Precio'])
        #Después de crear el df vacío lo guarda como un csv sin incluir indice mediante el uso de index=False
        df_productos.to_csv(archivo_productos, index=False)
        #Imprimo mensaje de éxito del proceso ejecutado
        print("\nDatos de productos borrados con éxito.")
    elif opcion_borrar == '2':
#En esta opción se crea un df vacío con las columnas cliente, fecha y producto que sobreescribe cualquier df anterior con datos de facturas
        df_facturas = pd.DataFrame(columns=['Cliente', 'Fecha', 'Productos'])
        #Luego de sobreescribir, guarda el nuevo df_facturas como un csv sin incluir indice, mediante el uso de index=False
        df_facturas.to_csv(archivo_facturas, index=False)
        #Imprimo mensaje de éxito del proceso ejecutado
        print("\nDatos de facturas borrados con éxito.")
    elif opcion_borrar == '3':
#En esta opción se crea un df vacío con las columnas nombre y precio que sustituye al df_productos anterior,
#También se crea un df vacío con las columnas cliente, fecha y productos que sustituye al df_facturas anterior
        df_productos = pd.DataFrame(columns=['Nombre', 'Precio'])
        df_facturas = pd.DataFrame(columns=['Cliente', 'Fecha', 'Productos'])
        #Imprimo mensaje de éxito del proceso ejecutado
        print("\nTodos los datos han sido borrados con éxito.")
#La última opción termina la ejecución de borrar_datos() y returna al usuario al menú principal
    elif opcion_borrar == '4':
        return


#-----------------------------------------------------------------------------------------
#F) Función principal del programa, con la opción salir
def main():
#Se utiliza la palabra clave global para indicar que se utilizarán las variables globales df_productos y df_facturas, para que la función main() acceda y pueda modificar estas variables
    global df_productos, df_facturas
#Creo el bucle principal del programa (utilizando while y if-elif-else) que comienza por mostrar el menú de opciones
    while True:
        #Imprimo el mensaje principal del menú
        print("\n----------------------------------------\nSistema de facturación y bodega")
        print("\nMenú Principal:")
        print("\na. Ingresar producto \nb. Ingresar facturas \nc. Ver resumen \nd. Ver detalle \ne. Borrar datos \nf. Salir")
        #Solicito al usuario que ingrese una opcion y la almaceno en la vairbale opcion
        #Además agrego el método lower() para convertir el texto ingresado a minúscula y conitnuar con el programa
        opcion = input("\nSeleccione una opción: ").lower()

        #Cada opción llamará a la función correspondiente para la acción deseada por el usuario
        if opcion == 'a':
            ingresar_producto()
        elif opcion == 'b':
            ingresar_factura()
        elif opcion == 'c':
            ver_resumen()
        elif opcion == 'd':
            ver_detalle()
        elif opcion == 'e':
            borrar_datos()
        elif opcion == 'f':
            #Imprimo un mensaje que le hace saber al usuario que eligió la opción de salir del programa
            print("\nEl programa finalizó. Adiós!")
            #Se agrega el break para terminar el bucle y finalizar la ejecución del programa
            break
        #Se agrega el else por si el usuario no apretó ninguna de las letras indicadas en el menú (a-f) e imprimo mensaje de error
        else:
            print("\nOpción no válida. Inténtelo de nuevo por favor.")

#Por último, se agrega esta condicional y seguidamente main(), para verificar que el script se esté ejecutando directamente como el programa principal
#Luego llama a la función principal main() para ejecutar el programa
if __name__ == "__main__":
    main()
```
