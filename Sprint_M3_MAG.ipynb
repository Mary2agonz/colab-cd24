#ACTIVIDAD N°5

---

Crear a partir del DataFrame una serie con los porcentajes de días que se ha conseguido el reto y los que no.


---
Respuesta:

```python
"""
import pandas as pd

#Creo el DataFrame
datos = {'Calorías': [420, 380, 390, 490, 300],
		    'Tiempo': [60, 40, 75, 55, 45]}

#En este nuevo df, datos es el diccionario que contiene los datos y para el index agregué la lista predefinida directamente en la función para simplificar el código
df = pd.DataFrame(datos, index= ['L', 'M', 'X', 'J', 'V'])

#Para calcular si se ha cumplido el reto, se divide el número de calorías por el tiempo(minutos) y se compara con 400 / 60 (400 calorías en una hora)
#Si el resultado de la división es mayor que 400/60, se asigna True, de lo contrario, False
df['reto'] = (df['calorias'] / df['tiempo']) > 400 / 60
"""
#Primero calculo el total de días evaluado por medio del método len()
total_dias=len(df)
#Para contar el número de días que cumplen y no cumplen el reto, comparo el resultado True y False, respectivamente; mediante el método sum() aplicado sobre la columna 'reto' del df
dias_cumplidos = (df['reto'] == True).sum()
dias_no_cumplidos = (df['reto'] == False).sum()

#Con las herramientas de división y multiplicacion calculo los porcentajes tomando en cuenta el total de días, y días cumplidos y no cumplidos resultantes del proceso previo
porcentaje_cumplidos = (dias_cumplidos / total_dias) * 100
porcentaje_no_cumplidos = (dias_no_cumplidos / total_dias) * 100

#Posterior a los calculos, creo la serie de porcentajes solicitada para el ejercicio
serie_porcentajes = pd.Series([f"{porcentaje_cumplidos: }%", f"{porcentaje_no_cumplidos: }%"], index=['Cumplidos', 'No cumplidos'])

# Mostrar la serie de porcentajes
print("El porcentaje de días que cumplieron y no el reto, se presenta a continuación: \n")
print(serie_porcentajes)
```

Crear un gráfico como el de más abajo que muestre la progresión de las calorías y tiempo durante la semana.


---
Respuesta:

```python
#import pandas as pd
#Importo la librería Matplotlib que llamaré con el alias 'plt'
import matplotlib.pyplot as plt

#Defino las variables que utilizaré para el gráfico
#los días quedarán determinados por el índice del df
dias = df.index
#Creo la variable calorías, que contiene los datos de la columna calorías del df
calorias = df['calorias']
#Creo la variable tiempo, que contiene los datos de la columna tiempo del df
tiempo = df['tiempo']

#Creo la figura para el gráfico, que será de 10 x 6 pulgadas de ancho y alto, respectivamente
plt.figure(figsize=(10,6))

#Agrego las líneas de código que representan a las calorías
#Se utiliza .scatter() y .plot() para trazar puntos y líneas en el gráfico, respectivamente
#alpha es la transparencia que se le aplicará a ese elemento, en este caso a la línea en el gráfico
plt.scatter(dias, calorias, color='purple', label='Calorías')
plt.plot(dias, calorias, color='purple', alpha=0.5)

#Agrego las líneas de código que representan al tiempo
#Se utiliza .scatter() y .plot() para trazar puntos y líneas en el gráfico, respectivamente
#Para el caso de los puntos de tiempo, cambié el marcador por forma de diamante
#alpha es la transparencia que se le aplicará a ese elemento, en este caso a la línea en el gráfico
plt.scatter(dias, tiempo, color='green', label='Tiempo', marker='D')
plt.plot(dias, tiempo, color='green', alpha=0.5)

#Agrego el título del gráfico
plt.title('Evolución de calorías y tiempo durante la semana')
#Agrego el nombre del eje x
plt.xlabel('Tiempo [minutos]')
#Agrego el nombre del eje y
plt.ylabel('Calorías [kcal]')
#Ubico la leyenda en la esquina superior izquierda y con un tamaño large
plt.legend(loc='upper left', fontsize='large')
#Agrego las líneas de la grilla del gráfico con estilo dotted que son puntos en vez de líneas
plt.grid(True, linestyle='dotted')
#Con la siguiente línea de código ajusto el gráfico para evitar superposiciones
plt.tight_layout()

#Finalmente muestro el gráfico
plt.show()
```
